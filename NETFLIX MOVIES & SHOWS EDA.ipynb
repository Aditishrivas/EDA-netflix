```python
netflix_titles_df['director'].fillna('No director',inplace = True) 
netflix_titles_df['cast'].fillna('No cast',inplace = True)
netflix_titles_df['country'].fillna('country unavailable',inplace = True)
netflix_titles_df.dropna(subset = ['date_added','rating','duration'],inplace = True)
```

```python
netflix_titles_df.isnull().any()
```

```python
netflix_movies_df = netflix_titles_df[netflix_titles_df['type']=='Movie'].copy() 
netflix_movies_df
```

```python
netflix_shows_df = netflix_titles_df[netflix_titles_df['type']=='TV Show'].copy() 
netflix_shows_df
```

```python
netflix_movies_df.duration = netflix_movies_df.duration.str.replace(' min','').astype 
netflix_shows_df.rename(columns={'duration':'seasons'},inplace=True)
netflix_shows_df.replace({'seasons':{'1 Season':'1 Seasons'}},inplace=True)
netflix_shows_df.seasons = netflix_shows_df.seasons.str.replace('Seasons','').astype
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(data=netflix_titles_df, x='type', palette='pastel')
plt.title("Count of Movies and TV Shows")
plt.xlabel("Type (Movies/TV Shows)")
plt.ylabel("Total Count")
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Create a pie chart
plt.pie(netflix_titles_df['type'].value_counts(), explode=(0.025, 0.025), labels=['Movies', 'TV Shows'], autopct='%1.1f%%', startangle=90)
plt.title('% of Netflix Titles that are either Movies or TV Shows')
plt.legend()
plt.show()
```

```python
order = ['G', 'TV-Y', 'TV-G', 'PG', 'TV-Y7', 'TV-Y7-FV', 'TV-PG', 'PG-13', 'TV-14', 'R', 'NC-17', 'TV-MA']
plt.figure(figsize=(15, 7))
sns.countplot(data=netflix_titles_df, x='rating', hue='type', order=order)
plt.title('Rating for Movies & TV shows')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
netflix_titles_df['year_added']=pd.DatetimeIndex(netflix_titles_df['date_added']).year
netflix_movies_df['year_added']=pd.DatetimeIndex(netflix_movies_df['date_added']).year 
netflix_shows_df['year_added']=pd.DatetimeIndex(netflix_shows_df['date_added']).year 
netflix_titles_df['month_added']=pd.DatetimeIndex(netflix_titles_df['date_added']).month
netflix_movies_df['month_added']=pd.DatetimeIndex(netflix_movies_df['date_added']).month
netflix_shows_df['month_added']=pd.DatetimeIndex(netflix_shows_df['date_added']).month
```

```python
# For Netflix titles DataFrame
netflix_year = netflix_titles_df['year_added'].value_counts().to_frame().reset_index().rename(columns={'year_added': 'count', 'index': 'year'})

# For Netflix movies DataFrame
movies_year = netflix_movies_df['year_added'].value_counts().to_frame().reset_index().rename(columns={'year_added': 'count', 'index': 'year'})

# For Netflix shows DataFrame
shows_year = netflix_shows_df['year_added'].value_counts().to_frame().reset_index().rename(columns={'year_added': 'count', 'index': 'year'})
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.lineplot(data=netflix_year, x='year', y='count', label='Netflix Titles')
sns.lineplot(data=movies_year, x='year', y='count', label='Movies')
sns.lineplot(data=shows_year, x='year', y='count', label='TV Shows')

ax.set_xticks(np.arange(2008, 2022, 1))
plt.xlabel("Year")
plt.ylabel("Count")

# Adding a legend
ax.legend()

plt.show()
```
